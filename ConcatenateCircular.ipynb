```python
class Node(object):
    def __init__(self,value):
        self.info=value
        self.link=None
```

```python
class CircularLinkedList(object):
    def __init__(self):
        self.last=None # Knowing last node in a circular list
        # makes insertion and deletion of element at the beginning and end of the list
        # possible in constant time and makes it suitable for implementing a queue
    def display_list(self):
        if self.last == None:
            print("Empty List")
            return 
    
#### Otherwise start at the last node and go to first node using its link
        p=self.last.link # this is first node! 
    
        while True:
            print(p.info," ",end='')
            p=p.link
            if p==self.last.link: # this means we traversed the whole list
                break  
        print()
    
    def insert_in_beginning(self,data):
        temp=Node(data)
        temp.link=self.last.link
        self.last.link=temp
        
    def insert_in_empty_list(self,data):   
        temp=Node(data)
        self.last=temp # Last and only node! 
        self.last.link=self.last # Refers to itself
        
    def insert_at_end(self,data):
        temp=Node(data)
        temp.link=self.last.link
        self.last.link=temp
        self.last=temp
        
    def create_list(self):
        n=int(input("Enter the number of nodes: "))
        
        if n==0: 
            return 
        
        data=int(input("Enter First Element to be inserted"))
        self.insert_in_empty_list(data)
        
        for i in range(n-1):
            data=int(input("Enter Next Element to be inserted"))
            self.insert_at_end(data)
      
    def insert_after(self,data,x):
        p=self.last.link
        while True:
            if p.info==x:
                break
            p=p.link
            if p==self.last.link:
                break
        ###Now you exited the while loop you either found or not found x 
        if p==self.last.link and p.info!=x: # exhausted the list and cannot find x
            print(x,"cannot be found in the list!")
        else: # you found x! 
            temp=Node(data)
            temp.link=p.link
            p.link=temp
            if p==self.last: # if p is last node , x will be last node i.e. replace p
                self.last=temp
                
    def delete_first_node(self):   
        if self.last is None:
            return
        
        if self.last.link==self.last: # one element! 
            self.last=None
            return
        self.last.link=self.last.link.link
    
    def delete_last_node(self): 
        if self.last is None:
            return
        if self.last.link==self.last:
            self.last=None
            return
        p=self.last.link
        while p.link!=self.last: # p refers to previous!
            p=p.link
        p.link=self.last.link
        self.last=p
        
               
    def delete_node(self,x): 
        if self.last is None:
            return
        
        if self.last.link==self.last and self.last.info==x: # only node!
            self.last=None
            return
        
        if self.last.link.info==x: # Delete first node!
            self.last.link=self.last.link.link
            return
        
        # otherwise initialize p to first node and traverse list
        p=self.last.link
        while p.link!=self.last.link:
            if p.link.info==x:
                break
            p=p.link
            
        if p.link==self.last.link:
            print(x,"not found in the list")
        else: #p refers to previous node
            p.link=p.link.link
            if self.last.info==x:
                self.last=p
                
    def concatenate(self,list2):
        # Trivial Cases
        if self.last is None:
            self.last=list2.last
            return
        if list2.last is None:
            return
        # Maintain reference to first element of list
        p=self.last.link
        self.last.link=list2.last.link
        list2.last.link=p
        self.last=list2.last
        
        
```

```python
## Client/Test Code for a circular list 

print("Create First List \n")
list1=CircularLinkedList()
list1.create_list()
list1.display_list()
print("Create Second List \n")
list2=CircularLinkedList()
list2.create_list()
list2.display_list()

list1.concatenate(list2)
print("Concatenated List \n")
list1.display_list()
```
