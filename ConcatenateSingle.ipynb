```python
class Node:
    def __init__(self,value):
        self.info=value
        self.link=None
```

```python
## Interface

class SingleLinkedList:
    def __init__(self):
        self.start=None
        
    def display_list(self):
        if self.start is None:
            print("Empty List")
        else:
            print("List is:  ")
            p=self.start
            while p is not None:
                print(p.info," ",end='')
                p=p.link
                
            print()
                
            
            
    def count_nodes(self):
        p=self.start
        n=0
        while p is not None:
                n+=1
                p=p.link
        print("Number of nodes in the list is= ",n )
    
    
    def search(self,x):
        p=self.start
        position=1
        while p is not None:
                if p.info==x:
                    print("Found element =",x," At position = ",position)
                    return True 
                
                p=p.link
                position+=1
        else:
            print("Could not find element =",x)
            return False
    
    def insert_in_beginning(self,data):
        temp=Node(data)
        temp.link=self.start
        self.start=temp
           
    def insert_at_end(self,data):
        temp=Node(data)
        # Case of empty list
        if self.start is None:
            self.start=temp
            return 
        ## Non-empty list, first traverse the list
        p=self.start
        while p.link is not None: 
            p=p.link
        #Now p is pointing to last node in the list
        p.link=temp
        
        
    ## Create list: 
    def create_list(self):
        n=int(input("Enter the number of nodes: "))
        if n==0: # Nothing to be done! 
            return 
        for i in range(n):
            data=int(input("Enter Element to be inserted"))
            self.insert_at_end(data)
            
################

    def insert_after(self,data,x):
        # Look for noe containing x 
        p=self.start
        while p is not None:
            if p.info==x:
                break
            p=p.link
        # If x is not found 
        if p is None:
            print(" Element ",x, "not found in the list")
        else: # Element found and now one needs to insert node
            temp=Node(data) # create a node
            temp.link=p.link
            p.link=temp
    
    def insert_before(self,data,x):
        # Check if list is empty
        if self.start is None:
            print("List is empty")
            return
        # if x is first element in a non-empty list (sort of trivial case!)
        if self.start.info==x:
            temp=Node(data) # create a node
            temp.link=self.start
            self.start=temp
            return 
        ## case x is not the first element , need to find reference to previous cell
        ## (i.e. predecessor of node containing x)
        p=self.start
        while p.link is not None:
            if p.link.info==x:
                break 
            p=p.link
        if p.link is None:
            print(x," Not present in the list")
        else: 
            temp=Node(data)
            temp.link=p.link
            p.link=temp

    def insert_at_position(self,data,k):
        if k==1: 
            temp=Node(data)
            temp.link=self.start
            self.start=temp
            return
        ## Need to find position before k 
        p=self.start
        i=1
        while i < k-1 and p is not None:
            p=p.link
            i+=1
        if p is None: ## outside list bounds
            print("you cannot insert beyond position index ",i)
        else:
            temp=Node(data)
            temp.link=p.link
            p.link=temp
        
            
    def delete_node(self,x):
        if self.start is None: #Case of Empty List
            print("Empty List!")
            return 
        # Case first node contain value x 
        if self.start.info==x: 
            self.start=self.start.link
            return 
        ## Traverse list up until predecessor of node containing x 
        p=self.start
        while p.link is not None:
            if p.link.info==x:  # Check if next cell contain x 
                break
            p=p.link
        if p.link is None:
            print("Element ",x, "is not in the list")
        else:
            p.link=p.link.link
    
    def delete_first_node(self):
        if self.start is None: # Empty list
            return
        self.start=self.start.link
    
    def delete_last_node(self):
        if self.start is None: # Empty list
            return
        if self.start.link is None: # Single Node List
            self.start=None
            return
        
        p=self.start # Traverse the list to the end 
        while p.link.link is not None: 
            p=p.link
        p.link=None
        
    
    
    def reverse_list(self):
        
        ## Initialize Previous, Current and Next pointers 
        prev=None
        p=self.start 
        ## traverse the list, break link to next and link back to previous 
        while p is not None: # p will be none in last iteration
             # capture next node (don't break the link to the list!)
            next=p.link
            # link back to previous node
            p.link=prev
            # move to next node (shif prev,p,next)
            prev=p
            p=next # This will exit the loop when p=None (= last node.link)
            ## At that point prev will be pointing to last node which is first node of the reverse
            ## list so now need to make this start of the reverse list
        self.start=prev
    
    def bubble_sort_exdata(self):
        # compare p,q starting at the beginning of the list and moving forward
        # swap if necessary
        # stop at the end. End here refers to the beginning of the sorted block of the list
        # at the beginning end refers to None, then last element after first pass...etc.
         
        end =None
        
        while end != self.start.link: # if end is the second node stop! 
            p=self.start # initialize p 
            
            while p.link!=end: # Move all the way to the beginning of the sorted block
                q=p.link
                if p.info > q.info: # compare
                    p.info,q.info=q.info,p.info  # Easy swap in Python!
                p=p.link
                
            end=p
            
                
    
    def bubble_sort_exlinks(self):
        ## Similar to Exchange data however this time exchange links only and rename p,q
        end =None
        
        while end != self.start.link: # if end is the second node stop! 
            r=p=self.start
            while p.link!=end: 
                q=p.link
                if p.info > q.info: # compare
                    p.link=q.link # let p points to node next to q 
                    q.link=p  # let q refer to node p 
                    # now need to let r refer to node q 
                    if p!=self.start: # Special case
                        r.link=q
                    else:
                        self.start=q # If first node and we did exchange self.start will need
                        # to link back to first node!
                        
                    p,q=q,p # Rename! 
                    
                ## Done Exchange now advance r,p,...
                r=p
                p=p.link
                
            end=p        
             
    def has_cycle(self): 
        if self.find_cycle() is None:
            return False
        else:
            return True
        
        
    def find_cycle(self):
        if self.start is None or self.start.link is None: #(list is empty or one node)
            return None # return reference to the node containing the cycle 
        
        slowR=self.start # Tortoise
        fastR=self.start # Hare
        
        while fastR is not None and fastR.link is not None:
            slowR=slowR.link
            fastR=fastR.link.link
            if slowR==fastR:
                return slowR
        return None
    
        
    def remove_cycle(self):
        c=self.find_cycle()
        if c is None:
            return 
        print("Node at which cycle is detected contains ",c.info)
        
        ## 
        p=c
        q=c
        length_cycle=0
        while True:
            length_cycle+=1
            q=q.link
            if p==q:
                break
        
        len_rem_list=0
        p=self.start
        while p!=q:
            len_rem_list+=1
            p=p.link
            q=q.link
        print("Number of nodes not included in the cycle are", len_rem_list)
        lenght_list=len_rem_list + length_cycle
        print("Length of the list is ",  lenght_list)
        
        
        p=self.start
        
        for i in range(lenght_list-1):
            p=p.link
            
        p.link=None
        
    
    def insert_cycle(self,x):
        if self.start is None:
            return
        p=self.start
        px=None
        prev=None
        
        while p is not None:
            if p.info==x:
                px=p
            prev=p
            p=p.link
        if px is not None:
            prev.link=px
        else:
            print(x,"Not in the List")
    
    
    def merge1(self,list2):
        merge_list=SingleLinkedList() ## This is a new empty
        #list; a placeholder for the merged list =(list1+list2) where list1 and list2 are
        # sorted lists.
        merge_list.start=self._merge1(self.start,list2.start) 
        # Function_merge1 combines list1 pointed to by self.start and list2 pointed to
        # by list2.start
        
    def _merge1(self,p1,p2):
        # p1 refers to first node in list1 and p2 points to first node of list2 initially
        if p1.info <= p2.info: 
            startM=Node(p1.info)
            p1=p1.link # advance p1 
        else:
            startM=Node(p2.info)
            p2=p2.link
            
        pM=startM
        
        while p1 is not None and p2 is not None: #(if either one becomes none exit!)
            if p1.info <= p2.info: 
                pM.link=Node(p1.info)
                p1=p1.link
            else:
                pM.link=Node(p2.info)
                p2=p2.link
            pM=pM.link
            
        ## Case when second list is finished and there are some elements left in first list
        while p1 is not None:
            pM.link=Node(p1.info)
            p1=p1.link
            pM=pM.link
        while p2 is not None:
            pM.link=Node(p2.info)
            p2=p2.link
            pM=pM.link
            
        return startM
    
    def merge2(self,list2):
        merge_list=SingleLinkedList() 
        merge_list.start=self._merge2(self.start,list2.start) 
        return merge_list
    
    def _merge2(self,p1,p2):
        if p1.info < p2.info:
            startM=p1
            p1=p1.link
        else:
            startM=p2
            p2=p2.link
        pM=startM
        
        while p1 is not None and p2 is not None:
            if p1.info < p2.info:
                pM.link=p1
                pM=pM.link
                p1=p1.link
            else:
                pM.link=p2
                pM=pM.link
                p2=p2.link
                
        if p1 is None:
            pM.link=p2
        else:
            pM.link=p1
        
        return startM
    
    
    #########
    def merge_sort(self):
        self.start=self._merge_sort_rec(self.start)
    def _merge_sort_rec(self,list_start):
        # Base condition, if list is empty or contain one element
        if list_start is None or list_start.link is None: 
            return list_start
        # For list containing more than one element call recursively! 
        start1=list_start
        start2=self._divide_list(list_start)
        start1=self._merge_sort_rec(start1)
        start2=self._merge_sort_rec(start2)
        startM=self._merge2(start1,start2)
        return startM
    
    def _divide_list(self,p):
        q=p.link.link
        while q is not None and q.link is not None:
            p=p.link # move p by one 
            q=q.link.link # move q by two 
        start2=p.link # beginning of second sublist
        p.link=None # end of first sublist
        return start2 
    
    def concatenate(self,list2):
        if self.start is None:
            self.start=list2.start
            return 
        if list2.start is None:
            return 
        
        p=self.start 
        while p.link is not None:
            p=p.link
            
        p.link=list2.start
    
        
        
```
