```python
class Node:
    def __init__(self,value):
        self.info=value
        self.link=None
```

```python
## Interface

class HeaderLinkedList:
    def __init__(self):
        self.head=Node(0) # alwasy there even for empty list. Value is initialized to 0 
        # but could be anything
        
    def display_list(self):
        if self.head.link is None:
            print("Empty List")
        else:
            print("List is:  ")
            p=self.head.link # p refers to first node in the list beyond the header
            
            while p is not None:
                print(p.info," ",end='')
                p=p.link
                
            print()
                
   ## Create list: 
    def create_list(self):
        # No need for special cases here 
        n=int(input("Enter the number of nodes: ")) 
        for i in range(n):
            data=int(input("Enter Element to be inserted"))
            self.insert_at_end(data)
            
    def insert_at_end(self,data):
        temp=Node(data)
   # No need for special cases
        p=self.head
        while p.link is not None: 
            p=p.link
        p.link=temp
        
    def insert_before(self,data,x):
   # No special cases compared to single linked list
   # only initialize at self.head instead of self.start
        p=self.head

        while p.link is not None:
            if p.link.info==x:
                break 
            p=p.link
        if p.link is None:
            print(x," Not present in the list")
        else: 
            temp=Node(data)
            temp.link=p.link
            p.link=temp

    def insert_at_position(self,data,k):
       ## Similarly no special conditions for k=1 like in single linked lists
        p=self.head
        
        i=1
        
        while i < k-1 and p is not None:
            p=p.link
            i+=1
        if p is None: ## outside list bounds
            print("you cannot insert beyond position index ",i)
        else:
            temp=Node(data)
            temp.link=p.link
            p.link=temp
       
    
################            
        
            
    def delete_node(self,x):
    
        p=self.head
        while p.link is not None:
            if p.link.info==x:  # Check if next cell contain x 
                break
            p=p.link
        if p.link is None:
            print("Element ",x, "is not in the list")
        else:
            p.link=p.link.link
        
    
    
    def reverse_list(self):
        
        ## Initialize Previous, Current and Next pointers 
        prev=None
        p=self.head.link
        
        ## traverse the list, break link to next and link back to previous 
        while p is not None: # p will be none in last iteration
             # capture next node (don't break the link to the list!)
            next=p.link
            # link back to previous node
            p.link=prev
            # move to next node (shif prev,p,next)
            prev=p
            p=next # This will exit the loop when p=None (= last node.link)
            ## At that point prev will be pointing to last node which is first node of the reverse
            ## list so now need to make this start of the reverse list
        self.head.link=prev

    
```

```python
## Client/Test Code for a singly-linked list 

list=HeaderLinkedList()
list.create_list()

while True:
    
    print("1. Display list")
    print("2. insert an element at the end of a list")
    print("3. insert a node before a specified node")
    print("4. insert a node at a specified position")
    print("5. delete a node")
    print("6. Reverse the list")
    print("7. Quit")
    
    
    option=int(input("Enter your Choice: "))
    
    if option==1:
        list.display_list()
    elif option==2:
        data=int(input("Enter the element to be inserted"))
        list.insert_at_end(data)
    elif option==3:
        data=int(input("Enter the element to be inserted"))
        x=int(input("Enter the element before which to insert"))
        list.insert_before(data,x)
    elif option==4:
        data=int(input("Enter the element to be inserted"))
        k=int(input("Enter the position at which to insert"))
        list.insert_at_position(data,k)
    elif option==5:
        data=int(input("Enter element to be deleted"))
        list.delete_node(data)
    elif option==6:   
        list.reverse_list()
    elif option==7:
        break
    else:
        print("Wrong Option")
    print()
```
