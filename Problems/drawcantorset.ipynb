```python
# Drawing lines between two points in python
import numpy as np
import matplotlib.pyplot as plt

class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
class line:
    def __init__(self,p1,p2):
        self.p1=p1
        self.p2=p2
       
        
    def plot(self):
        X=[self.p1.x , self.p2.x]
        Y=[self.p1.y, self.p2.y ]
        plt.plot(X,Y,"-ro")
        plt.xlabel("x")
        plt.ylabel("y")
     
p1=point(0,0)
p2=point(5,0)
line1=line(p1,p2)
line1.plot()
p3=point(0,5)
line2=line(p1,p3)
line2.plot()
```

```python
plt.figure(figsize=(12,8))


def drawline(startpoint,length):
    p2=point(startpoint.x+length,startpoint.y)
    line2=line(startpoint,p2)
    line2.plot()
    
def draw_cantor_set(startpoint,length, level):
    if level==1:
        drawline(startpoint,length)
    else:
        drawline(startpoint,length)
        p1=point(startpoint.x,startpoint.y-10)
        p2=point(startpoint.x+ (2./3.)*length , startpoint.y-10)
        draw_cantor_set(p1,length/3., level-1)
        draw_cantor_set(p2,length/3., level-1)
       
    
p1=point(0,0)
draw_cantor_set(p1,50,4)
```
